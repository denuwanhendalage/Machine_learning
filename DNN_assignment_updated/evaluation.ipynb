```python
# Evaluation Script for student
import os
import numpy as np
import csv
import zipfile


# ID = 'D:/CSEProjects/Machine_learning/DNN_assignment_updated/123456'  # Index number
ID = '123456'  # Index number

file_name = ['123456/dw_123456.csv', '123456/db_123456.csv'] # generated files
truth_path = ['true_dw_123456.csv','true_db_123456.csv'] # path of true files
threshold = 0.05


def read_file(name):
    l = []
    with open(name) as f:
        reader = csv.reader(f)
        for row in reader:
            cleaned_row = [np.float32(x) if x != '' else np.float32(0) for x in row]
            l.append(cleaned_row)
    return l


def read_file_from_zip(zip_path, filename):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        with zip_ref.open(filename) as f:
            reader = csv.reader(line.decode('utf-8') for line in f)
            data = []
            for row in reader:
                cleaned_row = [np.float32(x) if x != '' else np.float32(0) for x in row]
                data.append(cleaned_row)
    return data


def grade(l0, l1, th):
    dif = np.mean(np.abs(l0-l1).astype(float)/(0.1+l1))
    if dif <= th:
        return 1
    else:
        return 0

def compare(sub, true, threshold=0):
    scores = list()
    if not len(sub) == len(true):
        return 0
    for i in range(len(sub)):
        l0 = np.array(sub[i]).astype(float)
        l1 = np.array(true[i]).astype(float)
        if not len(l0) == len(l1):
            return 0
        else:
            scores.append(grade(l0, l1, threshold))
    return scores

true_grads = [read_file(f) for f in truth_path]


score = []
zip_file_path = ID + '.zip'

for i, fn in enumerate(file_name):
    grads = read_file_from_zip(zip_file_path, fn)
    s = compare(grads, true_grads[i], threshold)
    if isinstance(s, list):
        score += s
    else:
        score.append(s)

print("Total score:", np.sum(score)) # Correct answer will score you 122
# import zipfile

# zip_file_path = ID + '.zip'

# # List all files in the zip to see what's actually there
# with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
#     file_list = zip_ref.namelist()
#     print("Files in zip:", file_list)
```
